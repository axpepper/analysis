```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

import os, sys
```

```python
filename = 'BoxerCo.csv'
companyname = "BoxerCo"
```

```python
df_all = pd.read_csv(filename)
original = list(df_all.head(0))
changed = ['#', 'Organization', 'Position', 'Vision', 'comfort', 'Equipped', 'team', 'culture', 'purpose', 'success', 'inspiration', 'SD', 'FD', 'NID']
dic = {k:v for (k,v) in zip(original, changed)}
df_all.rename(dic, axis='columns')
```

```python
df_test = df_all[df_all["What is the name of your organization?"] == companyname]
```

```python
def anal_gen(df):
    for column_name in df:
        if df[column_name].dtype == 'int64':
            df[column_name].plot(kind = 'hist')
        if df[column_name].dtype == 'object':
            df[column_name].value_counts().plot(kind='bar')

        plt.show()
        #plt.savefig(column_name + ".png")
```

```python
anal_gen(df_all)
```
