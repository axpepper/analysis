```python
def anal_gen(df):
    for column_name in df:
        if df[column_name].dtype == 'int64':
            t = df[column_name].plot(kind = 'hist')
        if df[column_name].dtype == 'object':
            t = df[column_name].value_counts().plot(kind='bar')

        plt.show()
        plt.savefig(column_name + ".png")
```
